## Dim_Products

```python
df_products = spark.sql("SELECT ProductID,ProductNumber,ProductName,ProductModelName as ProductModel,ProductCategory,ProductSubCategory,ProductLine,Style,Color,SellStartDate,MakeFlag as ManufacturedInHouse FROM AdventureWorks_Lakehouse.Ops_Silver.products")
df_products.write.mode("overwrite").saveAsTable("Ops_Gold.Dim_Products")
```

## Dim_Calendar

```python
from pyspark.sql.functions import explode, sequence, to_date

beginDate = '2011-01-01'
endDate = '2015-12-31'

spark.sql(f"select explode(sequence(to_date('{beginDate}'), to_date('{endDate}'), interval 1 day)) as calendarDate").createOrReplaceTempView('dates')
```

```python
%%sql
select * FROM dates
```

```python
df_calendar = spark.sql(f'''
select
  year(calendarDate) * 10000 + month(calendarDate) * 100 + day(calendarDate) as DateInt,
  CalendarDate as Date,
  year(calendarDate) AS Year,
  date_format(calendarDate, 'MMMM') as Month,
  date_format(calendarDate,'MM') as MonthNo,
  quarter(calendarDate) as Quarter,
  CONCAT(string(year(calendarDate)),'-',date_format(calendarDate,'MM')) as YearMonth
from
  dates
order by
  calendarDate
''')
```

```python
df_calendar.write.mode("overwrite").saveAsTable("Ops_Gold.Dim_Calendar")
```
